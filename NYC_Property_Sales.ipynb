```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('nyc-rolling-sales.csv')
data.head()
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
print("Statistik deskriptif:\n", data.describe())
```

```python
plt.figure(figsize=(20, 8))
sns.histplot(data['YEAR BUILT'], bins=30, kde=True)
plt.xlabel('Harga Properti')
plt.ylabel('Frekuensi')
plt.title('Distribusi Harga Properti di NYC')
plt.show()
```

```python
data['YEAR BUILT'].plot.hist()
```

```python
mean_value = data['YEAR BUILT'].mean()
```

```python
grouped_data = data.groupby('YEAR BUILT').sum()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='YEAR BUILT', bins=20)
plt.title('Distribusi YEAR BUILT')
plt.xlabel('YEAR BUILT')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='TOTAL UNITS', y='YEAR BUILT')
plt.title('TOTAL UNITS Berdasarkan YEAR BUILT')
plt.xlabel('TOTAL UNITS')
plt.ylabel('YEAR BUILT')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='BLOCK', y='TAX CLASS AT TIME OF SALE')
plt.title('BLOCK vs. TAX CLASS AT TIME OF SALE')
plt.xlabel('BLOCK')
plt.ylabel('TAX CLASS AT TIME OF SALE')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap Korelasi antar Variabel')
plt.show()
```
